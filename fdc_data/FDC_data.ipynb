```python
import requests
import json
from config import api_key
```

```python
# New Dependency! Use this to pretty print the JSON
# https://docs.python.org/3/library/pprint.html
from pprint import pprint
```

```python
# Note that the ?t= is a query param for the t-itle of the
# movie we want to search for.
url = "https://api.nal.usda.gov/fdc/v1/foods/search?"
key = "api_key=" + api_key + "&query=FNDDS,"
fruit_name = "Apple"

query = (url+key+fruit_name)
print(query)
```

```python
# Performing a GET request similar to the one we executed
# earlier
foods = requests.get(query).json()
#data = response.json()
print(json.dumps(foods,indent=4,sort_keys=True))
```

```python
# Converting the response to JSON, and printing the result.
food_data = foods['foods'][0]
food_data
```

```python
fruit_name = food_data['description']
Publish_date = food_data['publishedDate']
Brand_owner = food_data['brandOwner']
print('fruit_name:' + fruit_name + ', ' + 'Publish_date:' + Publish_date + ', ' + 'Brand_owner:' + Brand_owner)
```

```python
fruit_name = foods['foods'][0]['description']
Publish_date = foods['foods'][0]['publishedDate']
Brand_owner = foods['foods'][0]['brandOwner']
print('fruit_name:' + fruit_name + ', ' + 'Publish_date:' + Publish_date + ', ' + 'Brand_owner:' + Brand_owner)
```

```python
nutrients = food_data['foodNutrients']
nutrients
```

```python
nID = []
nName = []
nNumber = []
uName = []
code = []
description = []
value = []

for x in nutrients:
    nid = x['nutrientId'] 
    nname = x['nutrientName']
    nnumber =x['nutrientNumber']
    unitname = x['unitName']
    dcode = x['derivationCode']
    desc = x['derivationDescription']
    val = x['value']
    nID.append(nid)
    nName.append(nname)
    nNumber.append(nnumber)
    uName.append(unitname)
    code.append(dcode)
    description.append(desc)
    value.append(val)
```

```python
import pandas as pd
fdc_data = pd.DataFrame()
```

```python
fdc_data['fruit_name'] = fruit_name
fdc_data['ID'] = nID
fdc_data['Name'] = nName 
fdc_data['Number'] = nNumber
fdc_data['Unit'] = uName
fdc_data['Code'] = code
fdc_data['Description'] = description
fdc_data['Value'] = value

fdc_data
```

```python
fdc_data.to_csv('fdc_data.csv')
```
